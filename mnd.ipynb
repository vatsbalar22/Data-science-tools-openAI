```python
import yfinance as yf
from datetime import datetime

def get_historical_stock_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data

# Example usage:
stock_symbol = "RELIANCE.BO"  # Replace with the desired stock symbol and .BO for BSE or .NS for NSE
start_date = "2000-01-01"  # Replace with the desired start date

# Set end_date to today's date
end_date = datetime.today().strftime('%Y-%m-%d')

historical_data = get_historical_stock_data(stock_symbol, start_date, end_date)

if not historical_data.empty:
    print(historical_data)
else:
    print("Failed to fetch historical data.")
```

```python
import pandas as pd

df = pd.DataFrame(historical_data)
df
```
